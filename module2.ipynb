```python
import pandas as pd 
data = pd.read_csv("week2.csv") 
data.set_index('Date')
```

```python
ax = plt.gca()
#ax.tick_params(direction='out', length=20, width=20, colors='r')
data.plot(kind='line',x='Date',y='Close Price',ax=ax,x_compat=True)
plt.show()
```

```python
x=data['Date']
y=data['Day_Perc_Change']
plt.stem(x,y,use_line_collection=True)
plt.show()
```

```python
date_name = "Date"
stem_name = "Day_Perc_Change"
plot_name = "No. of Trades"
t = data[date_name]
data1 = data[stem_name]
data2 = data[plot_name]

fig, ax1 = plt.subplots(figsize=(15,10))

color = 'tab:blue'
ax1.set_xlabel(date_name)
ax1.set_ylabel(stem_name, color=color)
ax1.stem(t, data1,use_line_collection=True)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:red'
ax2.set_ylabel(plot_name, color=color)  # we already handled the x-label with ax1
ax2.plot(t, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)
plt.xticks(rotation=90)
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

```python
exide = pd.read_csv("exide.csv")
voltas=pd.read_csv("voltas.csv")
mindtree=pd.read_csv("mindtree.csv")
jub=pd.read_csv("jubilant.csv")
castrol=pd.read_csv("castrol.csv")
```

```python
exide[exide["Series"]!="EQ"]
voltas[voltas["Series"]!="EQ"]
data=voltas['Close Price']
df = pd.DataFrame(data)
data=mindtree['Close Price']
df["mindtree"]=data
df.drop(['Close Price'],axis=1)
date=exide["Close Price"]
df["exide"]=exide["Close Price"]
df['castrol']=castrol["Close Price"]
df['dominos']=jub["Close Price"]
df['voltas']=voltas["Close Price"]
df['Date']=mindtree['Date']
df.set_index('Date')
df=df.drop(['Close Price'],axis=1)
df.head()
```

```python
pc=pd.DataFrame()
pc=df
pc=pc.drop(pc.index[0:])#dropping all rows
pc.at[0,'mindtree']=df.at[1,'mindtree']
for i in range(1,493):
    pc.at[i-1,'mindtree']=(df.at[i,'mindtree']-df.at[i-1,'mindtree'])/100
    pc.at[i-1,'exide']=(df.at[i,'exide']-df.at[i-1,'exide'])/100
    pc.at[i-1,'castrol']=(df.at[i,'castrol']-df.at[i-1,'castrol'])/100
    pc.at[i-1,'dominos']=(df.at[i,'dominos']-df.at[i-1,'dominos'])/100
    pc.at[i-1,'voltas']=(df.at[i,'voltas']-df.at[i-1,'voltas'])/100
pc
```
